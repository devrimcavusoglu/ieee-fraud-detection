```python
df = pd.read_parquet('./output/pre-modeling/all_num_data.pqt.gz')  
m, n = df.shape
print('Data Frame: {} x {}'.format(m,n))                
df.head()
```

```python
y = df.isFraud
df.drop(['isFraud', 'TransactionDT'], axis=1, inplace=True)
cols = df.columns
```

```python
desc = df.describe()
```

```python
log_cols = (cols[(cols.str.contains('C'))|(cols.str.contains('dist'))|(cols.str.contains('Transaction'))]).tolist()
```

```python
std_cols = list(set(df.columns.tolist()) - set(log_cols))
```

```python
log_cols, std_cols
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
num_log_scl = df.loc[:,log_cols]
std_cols = df.loc[:,std_cols]
del df
gc.collect()
```

```python
scaler = StandardScaler()
num_std_scl = pd.DataFrame(scaler.fit_transform(std_cols))
num_std_scl.shape
```

```python
num_log_scl = num_log_scl.apply(np.log,axis=0)
```

```python
num_log_scl.head()
```

```python
num_log_scl.replace([np.inf,-np.inf], np.nan, inplace=True)
num_log_scl.head()
```

```python
num_std_scl.columns = std_cols
num_std_scl.index = num_log_scl.index
```

```python
num_std_scl.head()
```

```python
num_log_scl.shape, num_std_scl.shape
```

```python
all_num_data = num_log_scl.join(num_std_scl, how='left')
```

```python
all_num_data.shape
```

```python
del num_log_scl, num_std_scl
gc.collect()
```
